```python
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold, GridSearchCV
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
print("done")
```

```python
abalone = pd.read_csv("datasets/abalone.csv")
print(abalone.head())
X = abalone.iloc[:, :-1]
y = abalone.Rings
```

```python
print(X)
print(y)
```

```python
X["Sex"] = X["Sex"].map({"M" : 1, "F" : -1, "I": 0})
```

```python
grid = {'n_estimators': range(1, 51)}
cv = KFold(n_splits=5, shuffle=True, random_state=1)
rfr = RandomForestRegressor(random_state=1)
gs = GridSearchCV(rfr, grid, scoring='r2', cv=cv, n_jobs=5)
gs.fit(X, y)
```

```python
gs.cv_results_
```

```python
for i, score in enumerate(gs.cv_results_["mean_test_score"]):
    if round(score, 2) > 0.52:
        print(gs.cv_results_["params"][i])
        break
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold, cross_val_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn import preprocessing
from sklearn.datasets import load_boston
```

```python
data, target = load_boston(return_X_y=True)
```

```python
data_scaled = preprocessing.scale(data)
linspace = np.linspace(1, 10, num=200, endpoint=True)
cv = KFold(n_splits=5, shuffle=True, random_state=42)
opt_power = -100000
max_score = -100000000
for power in linspace:
    clf = KNeighborsRegressor(n_neighbors=5, weights="distance", p=power, metric="minkowski")
    scores : np.ndarray = cross_val_score(clf, data_scaled, target, cv=cv, scoring="neg_mean_squared_error")
    mean_score = scores.mean()
    print(mean_score)
    if mean_score > max_score:
        opt_power = power
        max_score = mean_score

print(opt_power, max_score)
```

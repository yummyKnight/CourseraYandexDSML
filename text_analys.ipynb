```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn import datasets
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
print("done")
```

```python
newsgroups = datasets.fetch_20newsgroups(
                    subset='all',
                    categories=['alt.atheism', 'sci.space'])
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(newsgroups.data)
y = newsgroups.target
```

```python
print(X.shape)
print(y.shape)
```

```python
grid = {'C': np.power(10.0, np.arange(-5, 6))}
cv = KFold(n_splits=5, shuffle=True, random_state=241)
clf = SVC(kernel="linear", random_state=241)
gs = GridSearchCV(clf, grid, scoring='accuracy', cv=cv, verbose=2, n_jobs=5)
gs.fit(X, y)
gs.cv_results_
```

```python
print(gs.best_params_)
est =  gs.best_estimator_
```

```python
est.fit(X, y)
```

```python
s = pd.Series(abs(est.coef_.toarray()[0]))
print(s)
index_list = s.sort_values(ascending=False)[:10].index.tolist()
print("\n")
lex_list = [vectorizer.get_feature_names()[i] for i in index_list]
lex_list.sort()
print(*lex_list, sep=' ')
```
